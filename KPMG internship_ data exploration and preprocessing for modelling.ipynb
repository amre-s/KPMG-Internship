```python
# We perform fine tuning and coarse tuning of varaibles by calculating Weight of Evidence and finding Information Gain for 
#each split - 
## If weight of evidence is greater than 0 it is good predictor
#if information gain is high for a split then we conclude that coarse tuning is right
```

```python
# Function to find Weight of Evidence and information gain of each predictor
```

```python
def woe_discrete(df, discrete_variable_name, profit_binary):
        # we concate the values needed input and output variables
    df = pd.concat([df[discrete_variable_name], profit_binary], axis = 1)
        # take the count of number of values in each category and also the mean value which will give us the proortion of 1's in # # each category(as 1 represents good borrowers)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
     # now we calculate proportion of observation of each category
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

```python
# Function to plot the Weight of Evidence to find out which split gives high information gain
```

```python
def plot_by_woe(df_woe,rotation_of_x_axis_labels = 0):
    x = np.array(df_woe.iloc[: , 0].apply(str))
    y = df_woe['WoE']
    plt.figure(figsize=(18,8))
    plt.plot(x,y,marker = 'o',linestyle = '--', color = 'black')
    plt.xlabel(df_woe.columns[0])
    plt.ylabel('WoE')
    plt.title(str('Weight of evidence by ' + df_woe.columns[0]))
    plt.xticks(rotation = rotation_of_x_axis_labels)
```
